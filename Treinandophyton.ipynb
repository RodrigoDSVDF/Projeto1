```python
# bibliotecas necessárias
!pip install requests
!pip install beautifulsoup4

#  bibliotecas
import requests
from bs4 import BeautifulSoup

# URL da página
url = 'https://pt.wikipedia.org/wiki/Not%C3%ADcia'

# Fazendo uma requisição para a página
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida (código 200)
if response.status_code == 200:
    # Parseie o conteúdo da página com BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    #  realizar operações de scraping
    #  encontrar todos os links na página
    links = soup.find_all('a')

    # Imprima os links encontrados
    for link in links:
        print(link.get('href'))


else:
    print(f'Erro na requisição. Código: {response.status_code}')
```
